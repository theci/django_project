```python
import os # 경고 대응
os.environ['OMP_NUM_THREADS'] = '1'

import pandas as pd
import numpy as np

df = pd.read_csv("cal112_kind_of_average.csv")

df
```

```python
df1 = df.drop(['Unnamed: 0'], axis=1) 
df2 = df1[['지역','5대범죄 건수 평균', '인구밀도 평균', '여성 가구원수 평균', 'cctv수 평균', '가로등수 평균', '112신고 건수 평균']]
df2
```

```python
where = pd.get_dummies(df2['지역'])
df3 = df2.drop('지역', axis=1)
where
```

```python
df4 = df3.join(where)
df4
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn import linear_model

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
X = df4[df4.columns.difference(['112신고 건수 평균'])]
y = df4['112신고 건수 평균']

#검증(test)는 20%로 한다. = test_size=0.2
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=121)
```

```python

# 회귀 분석 객체 생성(선형 회귀 모델 생성)
lr = linear_model.LinearRegression()

#fit()는 기울기와 절편을 전달하기 위함.
model = lr.fit(X_train, y_train)

# 학습된 계수를 출력합니다.
print(lr.coef_)

# 상수항을 출력합니다.
print(lr.intercept_)
```

```python
y_predicted=model.predict(X)
print(y_predicted)
```

```python
df4["예측"] = y_predicted
df4
```

```python
df4[['112신고 건수 평균', '예측']]
```

```python
print(model.score(X_train, y_train)) # train R2 score를 출력합니다.
print(model.score(X_test, y_test)) # test R2 score를 출력합니다.
```
