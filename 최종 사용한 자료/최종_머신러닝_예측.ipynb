-- 각 파라미터 예측 --

```python
import pandas as pd

df=pd.read_csv('/content/data.csv')
```

```python
df
```

```python
regions=list(df['region'].unique())
regions
```

```python
df_ex=df[df['region'] == 'gangnam']
```

```python
df_ex
```

```python
df_expect=pd.DataFrame(columns=['cnt_crime','cnt_cctv','dense_population','cnt_light','cnt_police','female_population','cnt_notify','cnt_population'],index=regions)
```

```python
df_expect.head()
```

```python
from sklearn.linear_model import LinearRegression
import math

for r in regions:
  df_ex=df[df['region'] == r] # 특정구 df 추출

  X=df_ex['year'] # 년도 추출
  y_lists=['cnt_crime','cnt_cctv','dense_population','cnt_light','cnt_police','female_population','cnt_notify','cnt_population'] # 다양하게 뽑아낼 파라미터

  for l in y_lists: # 하나씩 예측해보자
    y=df_ex[l] # 특정구 하나의 파라미터 추출

    reg = LinearRegression().fit(X.values.reshape(-1,1), y) # 학습
    # reg.predict([[2022]]) # 예측

    print(r+'의 '+l+'는 '+'-> '+str(math.floor(reg.predict([[2022]]))))
    df_expect[l][r]=int(math.floor(reg.predict([[2022]])))
  print('\n')
```

```python
df_expect
```

```python
df_expect['region']=df_expect.index
df_expect.index=range(0,25)
df_expect
```

```python
df_past=pd.read_csv('/content/data.csv')
```

```python
df_past
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
df_past['region']=le.fit_transform(df_past['region'])
df_expect['region']=le.transform(df_expect['region'])
```

```python
df_past.head()
```

```python
df_expect.head()
```

```python
X=df_past.drop(['year','id','cnt_notify'],axis=1)
y=df_past['cnt_notify']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
import numpy as np

regr = LinearRegression()
regr.fit(X_train, y_train)

y_test_pred=regr.predict(X_test)
print(regr.score(X_test,y_test))

scores = cross_val_score(regr, X, y)
print(scores.mean())

print(mean_absolute_error(y_test, y_test_pred))
print(mean_squared_error(y_test, y_test_pred))
MSE = mean_squared_error(y_test, y_test_pred) 
print(np.sqrt(MSE))
```

```python
df_expect['cnt_notify']= 0
```

```python
df_expect['cnt_notify']=regr.predict(df_expect.drop(['cnt_notify'],axis=1))
```
